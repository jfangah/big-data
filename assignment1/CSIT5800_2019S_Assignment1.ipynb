## Step 4: Creating new features and dropping redundant features

### Step 4.0

### Step 4.1
Similarly, we can create more new features considering the followings:
<ul>
<li>
Overall quality of the house: product of OverallQual and OverallCond
</li><li>
Overall quality of the garage: product of GarageQual and GarageCond
</li><li>
Overall quality of the exterior: product of ExterQual and ExterCond
</li><li>
Overall kitchen score: product of KitchenAbvGr and KitchenQual
</li><li>
Overall fireplace score: product of Fireplaces and FireplaceQu
</li><li>
Overall garage score: product of GarageArea and GarageQual
</li><li>
Overall pool score: product of PoolArea and PoolQC
</li><li>
Total number of bathrooms: sum of BsmtFullBath, half of BsmtHalfBath, FullBath and half of HalfBath
</li><li>
Total SF for 1st + 2nd floors: sum of 1stFlrSF and 2ndFlrSF
</li><li>
Total SF for porch: sum of OpenPorchSF, EnclosedPorch, 3SsnPorch, and ScreenPorch
</li>
<ul>

```python
# Put your statements here
trainData["OverQualityOfHouse"]=trainData["OverallQual"] * trainData["OverallCond"]
trainData["OverQualituOfGarage"]=trainData["GarageQual"] * trainData["GarageCond"]
trainData["OverQualituOfExterior"]=trainData["ExterQual"] * trainData["ExterCond"]
trainData["OverKitchenScore"]=trainData["KitchenAbvGr"] * trainData["KitchenQual"]
trainData["OverFireplaceScore"]=trainData["Fireplaces"] * trainData["FireplaceQu"]
trainData["OverGarageScore"]=trainData["GarageArea"] * trainData["GarageQual"]
trainData["OverPoolScore"]=trainData["PoolArea"] * trainData["PoolQC"]
trainData["TotalBathrooms"]=trainData["BsmtFullBath"]+0.5*trainData["BsmtHalfBath"]+trainData["FullBath"]+0.5*trainData["HalfBath"]
trainData["TotalSFfor12"]=trainData["1stFlrSF"]+trainData["2ndFlrSF"]
trainData["TotalSFforPorch"]=trainData["OpenPorchSF"]+trainData["EnclosedPorch"]+trainData["3SsnPorch"]+trainData["ScreenPorch"]
```
